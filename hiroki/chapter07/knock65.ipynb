```python
sem_cnt = 0
sem_true = 0
syn_cnt = 0
syn_true = 0
with open(output, 'r', encoding='utf-8') as f:
    for row in f:
        cols = row.strip().split('\t')
        target = cols[1].split()[-1]
        pred = cols[2]
        if not cols[0].startswith('gram'):
            sem_cnt += 1
            if target == pred:
                sem_true += 1
        else:
            syn_cnt += 1
            if target == pred:
                syn_true += 1
print('意味的アナロジーの正解率: {}'.format(sem_true / sem_cnt))
print('文法的アナロジーの正解率: {}'.format(syn_true / syn_cnt))
```
