```python
def read_mecab_output(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        result = []
        for line in lines:
            if line.strip():
                surface, pos, base = line.strip().split('\t')
                result.append({'surface': surface, 'pos': pos, 'base': base})
        return result 
```

```python
def read_mecab(file):
    results = []
    morphs = []
    with open(file, 'r', encoding='utf-8') as f:
        for row in f:
            if row == 'EOS\n':
                if len(morphs):
                    results.append(morphs)
                    morphs = []
            else:
                cols1 = row.split('\t')
                cols2 = cols1[1].split(',')
                morph = {'surface': cols1[0],
                         'base': cols2[6],
                         'pos': cols2[0],
                         'pos1': cols2[1],
                        }
                morphs.append(morph)
    if len(morphs):
        results.append(morphs)
        morphs = []
    return results

results = read_mecab('Data/neko.txt.mecab')
results[:5]
```

```python
noun_list = []
for morphs in results:
    count = 0
    sent = ""
    for i in range(len(morphs)):
        if morphs[i]["pos"] == "名詞":
            count += 1
            sent += morphs[i]["surface"]
        else:
            if count >= 2:
              noun_list.append(sent)
            count = 0
            sent = ""
noun_set = set(noun_list)
noun_set
```
