```python
# モデルの構築
from torch import nn

class SLNet(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Linear(input_size, output_size)

    def forward(self, x):
        logits = self.fc(x)
        return logits

model = SLNet(300, 4)
print(model)
```
