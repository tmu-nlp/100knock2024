```python
from graphviz import Digraph
i = 0
for lines in kakari_list:
    graph = Digraph(format="png")
    graph.attr('edge', fontname = 'IPAGothic')
    for line in lines:
        text = line.split("\t")
        graph.node(text[0])
        graph.node(text[1])
        graph.edge(text[0], text[1])
    graph.render("[PATH]/output{}".format(str(i).zfill(3)))
    i += 1
```
