```python
class Morph:
    def __init__(self, txt):
        col1 = txt.split('\t')
        col2 = col1[1].rstrip().split(',')
        self.surface = col1[0]
        self.base = col2[6]
        self.pos = col2[0]
        self.pos1 = col2[1]

    def __str__(self):
        return 'surface: {}, base: {}, pos: {}, pos1: {}'.format(self.surface, self.base, self.pos, self.pos1)

    def __repr__(self):
        return self.surface


file_path = './ai.ja.txt.parsed'
morphs = []
morph = []
with open(file_path, 'r', encoding='utf-8') as f:
    for row in f:
        if row.startswith('*'):
            continue
        elif row == 'EOS\n':
            if len(morph):
                morphs.append(morph)
                morph = []
        else:
            morph.append(Morph(row))
if len(morph):
    morphs.append(morph)

for morph in morphs[1]:
    print(morph)
```
