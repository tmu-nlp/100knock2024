```python
import numpy as np
import pickle
import pandas as pd
from importnb import imports
with imports("ipynb"):
    import knock54 as k54

learning_rate = pickle.load(open("logistic_model.sav","rb"))
feat = k54.x_train.columns.values#len(feat) = 11032
zipped = zip(learning_rate.classes_,learning_rate.coef_)#classes_：数値の順番が確認できる．coef_:係数

for cat,coef in zipped:
    top = pd.DataFrame(feat[np.argsort(-coef)[:10]],\
        columns=["top10カテゴリ"],index = [i for i in range (1,11)])
    #argsort:並び替えたインデックスを返す(値は返さない)
    worst = pd.DataFrame(feat[np.argsort(coef)[:10]],\
        columns=["worst10カテゴリ"],index = [i for i in range (1,11)])
    print(f"{cat}の特徴量の重みランキング")
    print(top)
    print(worst)
    print("***************")
```
