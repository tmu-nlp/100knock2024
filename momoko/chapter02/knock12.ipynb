```python
import pandas as pd
df_names = pd.read_table("popular-names.txt",names=["名前","性別","人数","年"])

#1列目を抜き出す
f = open("col1.txt","w")
for name in df_names["名前"]:
    f.write(name+"\n")
f.close()

#2列目を抜き出す
f = open("col2.txt","w")
for sex in df_names["性別"]:
    f.write(sex+"\n")
f.close()
```
