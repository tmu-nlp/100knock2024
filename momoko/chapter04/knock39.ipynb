```python
"""
Zipfの法則とは
サイズの大きさで順位(k)を付けた場合，k番目のサイズは，1番目のサイズの1/kになるという経験則．
対象テキストにおける語の出現頻度が↑に従うことが発見されている．
"""
from importnb import imports
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import collections

with imports("ipynb"):#これで"knock35.~~~"の形でknock35内で定義したものを再利用可能となった
    import knock35

data = collections.Counter(knock35.neko_words)
logarithm = sorted((data.values()),reverse = True)#逆順にソート
plt.plot(logarithm)
plt.xlabel("出現頻度順位")
plt.ylabel("単語出現頻度")
plt.title("Zipfの法則")
plt.yscale('log')#y軸をlogスケールで描く
plt.xscale('log')
plt.show()
```
