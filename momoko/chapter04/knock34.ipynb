```python
from importnb import imports
with imports("ipynb"):#これで"knock30.~~~"の形でknock30内で定義したものを再利用可能となった
    import knock30

#surface(表層形)，base(基本形)，pos(品詞)，pos1(品詞細分類1)
for l in knock30.neko_dict:
    meisi_tati=""
    num = 0
    for i in range (len(l)):
        if l[i]["pos"]=="名詞":
            meisi_tati += l[i]["surface"]
            num += 1
        else:
            if num > 2:#連続して名詞が出現していた場合は
                print(meisi_tati)
            meisi_tati=""
            num = 0
```
