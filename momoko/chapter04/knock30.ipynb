```python
import MeCab
basepath = "/Users/shirakawamomoko/Desktop/nlp100保存/chapter04/"
with open (basepath+"neko.txt.mecab","r")as f:
    lines = f.readlines()
    neko_dict=[]#1文ごとの形態素(=辞書にしたもの)のリスト
    sentence_list=[]#1文ごとに格納する準備のリスト
    for l in lines:
        l = l.replace("\n","").split("\n")[0]#解析結果ごとのリスト["すもも\tスモモ\tスモモ\t李\t名詞-普通名詞-一般"]
        minil = l.split("\t")#["すもも","スモモ","スモモ","李","名詞-普通名詞-一般"]
        if minil[0]=="EOS":
            neko_dict.append(sentence_list)
            sentence_list=[]
            continue
        if len(minil)==0 or minil[0]=="\u3000":
            continue
        dict={}
        dict["surface"]=minil[0]#表層形
        dict["base"]=minil[3]#基本形
        dict["pos"]=minil[4].split("-")[0]#品詞
        try:
            dict["pos1"]=minil[4].split("-")[1]#品詞細分類1
        except IndexError:#もし品詞細分類1がなければ
            dict["pos1"]="なし"
        sentence_list.append(dict)
        dict={}
```
