```python
from importnb import imports
import collections
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

with imports("ipynb"):#これで"knock30.~~~"の形でknock30内で定義したものを再利用可能となった
    import knock30

neko_co = []
#surface(表層形)，base(基本形)，pos(品詞)，pos1(品詞細分類1)
for l in knock30.neko_dict:
    counter = 0
    for keitaiso in l:
        if "猫" in keitaiso["surface"]:#猫が1文に含まれていた場合
            counter = 1
        if keitaiso["pos"]=="補助記号" or keitaiso["pos"]=="助詞" or keitaiso["pos"]=="助動詞":
            continue
        if counter == 1:#counterが1の時だけ
            neko_co.append(keitaiso["surface"])

neko_co_frequency = collections.Counter(neko_co).most_common()[::1]
neko_co_frequency

#グラフ作成
neko_f_top10 =neko_co_frequency[1:11]#1位が「猫」だったので含まない
top10_num = []
top10_rabel = []
for atai in neko_f_top10:
    top10_rabel.append(atai[0])
    top10_num.append(atai[1])

plt.bar(top10_rabel,top10_num)
plt.xlabel("形態素")
plt.ylabel("出現頻度(回)")
plt.title("『吾輩は猫である』における，「猫」と共起頻度の高い形態素top10")
plt.show()
```
