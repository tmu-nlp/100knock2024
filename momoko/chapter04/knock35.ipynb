```python
from importnb import imports
import collections

with imports("ipynb"):#これで"knock30.~~~"の形でknock30内で定義したものを再利用可能となった
    import knock30

neko_words = []
#surface(表層形)，base(基本形)，pos(品詞)，pos1(品詞細分類1)
for l in knock30.neko_dict:
    for keitaiso in l:
        if keitaiso["pos"]=="補助記号" or keitaiso["pos"]=="助詞" or keitaiso["pos"]=="助動詞":
            continue
        neko_words.append(keitaiso["surface"])

#collections.Counter：キーに単語，値に出現回数(collections.Counter)
#most_common()[::1]：出現回数が多い順に並び替え(list)
neko_frequency = collections.Counter(neko_words).most_common()[::1]
print(neko_frequency)
```
