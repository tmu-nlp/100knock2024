```python
from importnb import imports

with imports("ipynb"):
    import knock41

for sentence in knock41.ai_phrase_dict:
    for noun_kana in sentence:
        if "名詞" in noun_kana["pos"]:#もし名詞が入っていたら
            for v in sentence:
                if noun_kana["dst"] in v["srcs"]:#もしnoun_kanaのdstがsrcsにいたら
                    verb_kana = v
                    if "動詞" in verb_kana["pos"]:#もし動詞が入っていたら
                        clause_noun = "".join(noun_kana["morphs"][no] if noun_kana["pos"][no] != "記号" else "" for no in range (len(noun_kana["morphs"])))#記号でなければmorphを連結
                        clause_verb = "".join(verb_kana["morphs"][ve] if verb_kana["pos"][ve] != "記号" else "" for ve in range (len(verb_kana["morphs"])))
                        print(f"{clause_noun}\t{clause_verb}")
```
