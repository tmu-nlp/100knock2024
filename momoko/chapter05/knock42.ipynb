```python
from importnb import imports
with imports("ipynb"):
    import knock41 as k41

def tabu_kugiri(file_name):#file_name:f.readlines()のやつ
    chunks = []#全体の文字保存
    chunk_save = []#chunksのための一時保存先
    links = []
    for line in file_name:
        if line[0] == "*":
            if chunk_save:#もし文字列がいれば
                chunks.append(''.join(chunk_save))#文字列を連結
                chunk_save = []
            _, chunk_id, link, *rest = line.split()
            #_：無視する　chunk_id：文節番号　link：係り先の文節番号(D付き)　*rest：それ以外
            links.append((int(chunk_id), int(link.replace("D",""))))
        elif line.strip() == 'EOS':
            if chunk_save:
                chunks.append(''.join(chunk_save))#文字列を結合してchunksに保存
            if chunks:
                for chunk_id, link in links:
                    if link != -1:#係り先の文節番号が-1でなければ
                        print(f'{chunks[chunk_id]}\t{chunks[link]}')#文節番号，係り先分析番号に対応する文字列をprint
            chunks = []
            chunk_save = []
            links = []
        else:
            word= line.split('\t')[0]
            pos = line.split('\t')[1].split(",")[0]#品詞(記号じゃないかの判定に使う)
            if pos != "記号":#もし記号でなければ入れる
                chunk_save.append(word)


basepath = "/Users/shirakawamomoko/Desktop/nlp100保存/chapter05/"
with open(basepath + "ai.ja.txt.parsed", "r") as f:
    file_name = f.readlines()
tabu_kugiri(file_name)

```
