```python
#参考：https://qiita.com/g-k/items/120f1cf85ff2ceae4aba
from importnb import imports
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

with imports("ipynb"):
    from knock67 import *
```

```python
tsne = TSNE(n_components = 2,random_state=20030220)#n_components:2次元空間
vec_num_numpy = np.array(vec_num)
country_embedded = tsne.fit_transform(vec_num_numpy)
pred = KMeans(n_clusters=5,random_state=20030220).fit_predict(vec_num_numpy)
```

```python
col_list = ["Blue","Green","Yellow","Pink"]
plt.figure(figsize=(10,10))
for tsne,name,km in zip(country_embedded,vec_country,pred):
    plt.plot(tsne[0],tsne[1],color= col_list[km-1],marker="o")#tsneから座標を指定，色はkmのクラスタから指定．形をoにする．
    plt.annotate(name,xy=(tsne[0],tsne[1]))#国名を点ごとにつける
plt.title("knock69_T-SNE")
plt.savefig("knock69.png")
```
