```python
#21 記事中でカテゴリ名を宣言している行を抽出せよ．
```

```python
import pandas as pd
import re
pattern = re.compile('Category') #re.compile(pattern)で特定の文字列パターンをオブジェクトとして生成
wiki = pd.read_json('jawiki-country.json.gz', lines = True)
uk = wiki[wiki['title']=='イギリス'].text.values
ls = uk[0].split('\n')
for line in ls:
    if re.search(pattern, line):
        print (line)
```
