```python
#24 記事から参照されているメディアファイルをすべて抜き出せ．
#メディアファイル=ウィキペディア内の画像、音声、動画などのメディアファイルに関する最上位カテゴリ
```

```python

import pandas as pd
import re
pattern = re.compile('File|ファイル:(.+?)\|') #'File''ファイル:' の後に続く任意の文字列で、'|' で終わるもの
wiki = pd.read_json('jawiki-country.json.gz', lines = True)
uk = wiki[wiki['title']=='イギリス'].text.values
ls = uk[0].split('\n')
for line in ls:
    r = re.findall(pattern, line)
    if r:
        print (r[0])
```
