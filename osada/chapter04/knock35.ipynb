```python
#35 文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．
```

```python
from collections import Counter

result = []
sentence = []

with open("neko.txt.mecab") as f:
  for line in f:
    l1 = line.split("\t")
    if len(l1) == 2:
      l2 = l1[1].split(",")
      sentence.append({"surface": l1[0], "base": l2[6], "pos": l2[0], "pos1": l2[1]})
      if l2[1] == "句点":
        result.append(sentence)
        sentence = []

text_l = []

for lis in result:
  for dic in lis:
    if dic["surface"] != "":
      text_l.append(dic["surface"])

c35 = Counter(text_l)
print(c35.most_common())
```
