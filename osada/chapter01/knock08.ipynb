```python
##8 与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．
#英小文字ならば(219 - 文字コード)の文字に置換
#その他の文字はそのまま出力
#この関数を用い，英語のメッセージを暗号化・復号化せよ．

def cipher(t8):
  newt8=""
  for i in t8:
    if i.islower() and i.isalpha():
      i=chr(219-ord(i)) #.islower 小文字 isalpha　アルファベット
    newt8+=i
  return newt8

print(cipher("I am an NLPer"))
```
