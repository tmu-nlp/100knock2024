```python
#16　自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．
```

```python
N = 3
import pandas as pd
df = pd.read_csv('popular-names.txt', delimiter='\t', header=None) #delimeter= "\t"でタブを区切り文字として認識
step = - (-len(df) // N)
for n in range(N):
    df_split = df.iloc[n*step:(n+1)*step]
    df_split.to_csv('popular-names'+str(n)+'.txt', sep='\t',header=False, index=False)
```
