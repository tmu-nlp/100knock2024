```python
#No72(損失と勾配の計算)

#pklファイルを読み込むためのクラス定義(ファイルにはクラスの情報も入っているため)
class SLP(nn.Module):
    def __init__(self, input_size, output_size):
      #親クラスのコンストラクタの呼び出し
        super().__init__()
        self.fc = nn.Linear(input_size, output_size, bias=False)
        # 正規乱数(平均0,標準偏差1の確率分布から乱数を取り出す）で重みを初期化する
        nn.init.normal_(self.fc.weight, mean=0.0, std=1.0)
#順伝播：　入力層から出力層への方向を順方向として入力変数とパラメータをかけ合わせて予測値を計算すること
    def forward(self, x):
        x = self.fc(x)
        return x
```

```python
#交差エントロピー損失は，負の対数尤度損失と考えるとわかりやすい(尤もらしくない事例の損失が大きくなる)
# 勾配：　関数の傾きや変化率のこと。最小値または最大値の方向を指し示す指標
#2次関数の最適化(最小値、最大値）問題の例で考えるとわかりやすい
import torch
from torch import nn
import pickle
#ファイル、学習データ、検証データの読み込み
model = pickle.load(open("/content/drive/MyDrive/SLP.pkl", "rb"))
X_train = torch.load("/content/drive/MyDrive/X_train.pt")
y_train = torch.load("/content/drive/MyDrive/y_train.pt")
#損失関数のインスタンスの作成
criterion = nn.CrossEntropyLoss()

# l1のクロスエントロピー誤差
l1 = criterion(model(X_train[0]), y_train[0])
model.zero_grad()
#誤差逆伝播をして、勾配をもとめる
l1.backward()
print("クロスエントロピー損失:", l1.item())
print("勾配", model.fc.weight.grad)

# l1からl4のクロスエントロピー誤差
l = criterion(model(X_train[:4]), y_train[:4])
#パラメータの初期化
model.zero_grad()
l.backward()
print("クロスエントロピー損失:", l.item())
print("勾配", model.fc.weight.grad)
```
