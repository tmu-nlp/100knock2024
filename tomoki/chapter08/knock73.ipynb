```python
#No73(確率的勾配降下法による学習)
#確率的勾配降下法のよいところは、ランダム性によって局所最適解に陥りづらいところや、収束が早いところにある
#データセットとデータローダ(データセットの中身をミニバッチごとに固めた集合)の作成に必要
import torch
from torch.utils.data import TensorDataset, DataLoader
#モデルの作成
class LogisticRegression(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.net = torch.nn.Sequential(
            torch.nn.Linear(300, 4),
        )
    def forward(self, X):
        return self.net(X)

model = LogisticRegression()
X_train = torch.load("/content/drive/MyDrive/X_train.pt")
y_train = torch.load("/content/drive/MyDrive/y_train.pt")
#TensorDatasetを作成
ds = TensorDataset(X_train, y_train)
# DataLoaderを作成
loader = DataLoader(ds, batch_size=1, shuffle=True)
#損失
loss_fn = torch.nn.CrossEntropyLoss()
#困ったら学習率は0.001にするとよい(おそらく有名論文の結果に基づいていると思われる)
#結果を見た感じ、今回は0.1くらいでよかったかもしれない
optimizer = torch.optim.SGD(model.net.parameters(), lr=1e-3)
#エポック数100で学習時間はおよそ10分
epochs=100
for epoch in range(epochs):
    for xx, yy in loader:
        y_pred = model(xx)
        loss = loss_fn(y_pred, yy)
        #勾配の初期化
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
#モデルの保存
torch.save(model.state_dict(), "/content/drive/MyDrive/knock73.pth")
```
