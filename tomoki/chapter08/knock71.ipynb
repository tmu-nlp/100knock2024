```python
#No71(単層ニューラルネットワークによる予測)
#単層ニューラルネットワーク(入力層と、出力層からなる)
#シグモイド関数　→　2値分類(〇か✕か)をしたいときに使う
#ソフトマックス関数　→　多値分類をしたいときに使う(0以上1以下の値が出力される、出力の合計値が1になる)
import torch
from torch import nn
import pickle
#「torch.nn.Moduleクラスのサブクラス（子クラス）化」によるモデルの定義
class SLP(nn.Module):
    def __init__(self, input_size, output_size):
      #親クラスのコンストラクタの呼び出し
        super().__init__()
        self.fc = nn.Linear(input_size, output_size, bias=False)
        # 正規乱数(平均0,標準偏差1の確率分布から乱数を取り出す）で重みを初期化する
        nn.init.normal_(self.fc.weight, mean=0.0, std=1.0)
#順伝播：　入力層から出力層への方向を順方向として入力変数とパラメータをかけ合わせて予測値を計算すること
    def forward(self, x):
        x = self.fc(x)
        return x


X_train = torch.load("/content/drive/MyDrive/X_train.pt")
# 入力チャネル数(inputsize)と出力チャネル数(outputsize)の指定
model = SLP(300, 4)
with open('SLP.pkl', 'wb') as f:
    pickle.dump(model, f)

#softmax関数の利用(列単位での計算)
y_hat_1 = torch.softmax(model(X_train[0]), dim=0)
print(y_hat_1)
#softmax関数の利用(行単位での計算)
Y_hat = torch.softmax(model.forward(X_train[:4]), dim=1)
print(Y_hat)
```
