```python
import torch
from torch.utils.data import TensorDataset, DataLoader
#モデルの作成
class LogisticRegression(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.net = torch.nn.Sequential(
            torch.nn.Linear(300, 4),
        )
    def forward(self, X):
        return self.net(X)

model = LogisticRegression()
model.load_state_dict(torch.load("/content/drive/MyDrive/knock73.pth"))
#訓練データセットの作成
model = LogisticRegression()
X_train = torch.load("/content/drive/MyDrive/X_train.pt")
y_train = torch.load("/content/drive/MyDrive/y_train.pt")
ds = TensorDataset(X_train, y_train)
train_dataloader = DataLoader(ds, batch_size=1, shuffle=True)
```

```python
# テストデータセットの作成
X_test = torch.load("/content/drive/MyDrive/X_test.pt")
y_test = torch.load("/content/drive/MyDrive/y_test.pt")
ds = TensorDataset(X_test, y_test)
test_dataloader = DataLoader(ds, batch_size=1, shuffle=True)

def calculate_accuracy(model, loader):
    model.eval()
    total = 0
    correct = 0
    with torch.no_grad():
        for inputs, labels in loader:
            outputs = model(inputs)
            pred = torch.argmax(outputs, dim=-1)
            total += len(inputs)
            correct += (pred == labels).sum().item()
    return correct / total


train_accuracy = calculate_accuracy(model, train_dataloader)
test_accuracy = calculate_accuracy(model, test_dataloader)
#73で作成したモデルの学習率が低すぎたのが反省点(学習率が0.1だと正解率が0.9を超えた)
print(f'train_accuracy : {train_accuracy:.3f}')
print(f'test_accuracy : {test_accuracy:.3f}')

#train_accuracy : 0.467
#test_accuracy : 0.465


```
