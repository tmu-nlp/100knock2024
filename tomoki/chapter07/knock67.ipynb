```python
#No67(k-meansクラスタリング)
from gensim.models import keyedvectors
import pickle
import pandas as pd
import numpy as np
#k-meansのライブラリをimport
from sklearn.cluster import KMeans

with open('/content/drive/MyDrive/GoogleNews-vectors.pkl', 'rb') as f:
    model = pickle.load(f)

countries = []
countries_name = []
vec_countries = []
#_sigをつけるかどうかは、場合による(文字化け対策になることがある)
with open('/content/drive/MyDrive/countries_name.csv', 'r', encoding='utf-8_sig') as f:
    for line in f:
      # 改行を削除
        country = line.rstrip()
        #リストに追加
        countries.append(country)
 # ヘッダの削除(指定した値を削除できる.また削除した値を取得できる)
countries.pop(0)

# モデルに含まれる国名のみを抽出(try except でも可)
for country in countries:
    if country in model:
        countries_name.append(country)
        vec_countries.append(model[country])  # 単語ベクトルの取得

# k-meansモデル作成(分類するごとに、単語ベクトルの重心を求めなおし、重心が定まったとき、実行が終わる)
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)  # n_clusters..クラスタ数(デフォルト8)

# クラスタリング実行
kmeans.fit(vec_countries)

for i in range(k):
  #whereを利用することによって、数字のラベルを、clusterに入れることができる。
    cluster = np.where(kmeans.labels_ == i)[0]
    print(f'cluster:{i}')
    print(", ".join([countries[j] for j in cluster]))
    print('-'*100)
```
