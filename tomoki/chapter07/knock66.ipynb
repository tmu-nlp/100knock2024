```python
#No66(WordSimilarity-353での評価)
import pickle
import pandas as pd
#ファイルの読み込み
with open('/content/drive/MyDrive/GoogleNews-vectors.pkl', 'rb') as f:
    model = pickle.load(f)
#csvファイルのdf化
df = pd.read_csv('/content/drive/MyDrive/combined.csv')
#dfの行数分だけ繰り返す。ilokで1行ずつ取得していく。
#lineの「Word1」と「Word2」の列をsimに追加する。
sim = []
for i in range(len(df)):
    line = df.iloc[i]
    sim.append(model.similarity(line['Word 1'], line['Word 2']))
#w2vの列にsimを追加する
df['w2v'] = sim
#確認
print(df['w2v'])
#スピアマンの相関係数を求める。(他にはピアソンやケンドールの順位相関係数もある)
result = df[['Human (mean)', 'w2v']].corr(method='spearman')
print(f'spearman corr: {result}')
```
