```python
#No65(アナロジータスクでの正解率)
#semantic analogy 意味的に近いか　#syntactic analogy 文法的に近いか
with open('/content/drive/MyDrive/txt.64', 'r', encoding='utf-8_sig') as f:
    sem_cnt = 0  # 意味的アナロジー(semantic analogy)分母
    sem_cor = 0  # 意味的アナロジー分子
    syn_cnt = 0  # 文法的アナロジー(syntactic analogy)分母
    syn_cor = 0  # 文法的アナロジー分子

    for line in f:
      # 「カテゴリ | 評価データ | 単語, 類似度」を | 区切りで3分割
        line = line.split('|')
        category = line[0]
        #評価データの分割
        data = line[1].split()
        # 単語と類似度に分割
        word, cos = line[2].split()

        # 文字列(カテゴリ)がgramで始まるかをstartswithで検査(特定のフレーズで始まるかどうか)
        #正解していれば分母と分子が増えるが、間違えると分母だけ増えてしまい、正解率が落ちる
        if not category.startswith('gram'):
            sem_cnt += 1
            if data[3] == word:  # knock64の予測が正解しているかどうか
                sem_cor += 1
        else:
            syn_cnt += 1
            if data[3] == word:
                syn_cor += 1
# 意味的アナロジーの正解率を計算
sem_accuracy = sem_cor / sem_cnt
# 文法的アナロジーの正解率を計算
syn_accuracy = syn_cor / syn_cnt

print('Accuracy(semantic analogy) : ', sem_accuracy)
print('Accuracy(syntactic analogy) : ', syn_accuracy)
```
