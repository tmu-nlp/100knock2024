```python
import pandas as pd
import joblib

# モデルとエンコーダーを読み込む
model = joblib.load('logistic_regression_model.pkl')
label_encoder = joblib.load('label_encoder.pkl')
vectorizer = joblib.load('vectorizer.pkl')

# 特徴量の重みを取得
feature_names = vectorizer.get_feature_names_out()
coefs = model.coef_

# 重みが最も高いトップ10と低いトップ10の特徴量を確認
for i, label in enumerate(label_encoder.classes_):
    print(f"Top 10 positive features for {label}:")
    top10 = coefs[i].argsort()[-10:][::-1]
    print("\n".join([f"{feature_names[j]}: {coefs[i][j]:.4f}" for j in top10]))
    
    print(f"\nTop 10 negative features for {label}:")
    bottom10 = coefs[i].argsort()[:10]
    print("\n".join([f"{feature_names[j]}: {coefs[i][j]:.4f}" for j in bottom10]))
    print("\n")
```
