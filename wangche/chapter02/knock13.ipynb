12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```python
with open("col1.txt", "r") as col1, open("col2.txt", "r") as col2, open("merged.txt", "w") as merged:
    col1_lines = col1.readlines()
    col2_lines = col2.readlines()
    for col1_line, col2_line in zip(col1_lines, col2_lines):
        merged.write(col1_line.strip() + "\t" + col2_line.strip() + "\n")
```

```python
with open("merged.txt", "r") as f:
    for line in f:
        print(line.strip())
```
