名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．


```python
def extract_noun_sequences():
    noun_sequences = []

    with open('neko.txt.mecab', 'r', encoding='utf-8') as infile:
        sentence = []
        for line in infile:
            if line == 'EOS\n':
                # 文の中から名詞の連接を抽出
                i = 0
                while i < len(sentence):
                    if sentence[i]['pos'] == '名詞':
                        noun_sequence = sentence[i]['surface']
                        i += 1
                        while i < len(sentence) and sentence[i]['pos'] == '名詞':
                            noun_sequence += sentence[i]['surface']
                            i += 1
                        noun_sequences.append(noun_sequence)
                    else:
                        i += 1
                sentence = []
                continue
            
            # 分割表面形とその他の属性
            surface, attr = line.split('\t')
            attr = attr.split(',')
            morph = {
                'surface': surface,
                'base': attr[6],
                'pos': attr[0],
                'pos1': attr[1]
            }
            sentence.append(morph)

    return noun_sequences

```

```python
if __name__ == '__main__':
    noun_sequences = extract_noun_sequences()
    for sequence in noun_sequences:
        print(sequence)
```
