```python
#knock80
#80. ID番号への変換
#問題51で構築した学習データ中の単語にユニークなID番号を付与したい．
#学習データ中で最も頻出する単語に1，2番目に頻出する単語に2，……といった方法で，
#学習データ中で2回以上出現する単語にID番号を付与せよ．
#そして，与えられた単語列に対して，ID番号の列を返す関数を実装せよ．
#ただし，出現頻度が2回未満の単語のID番号はすべて0とせよ．

import pandas as pd
from collections import defaultdict
import string

#与えられたデータセット内の単語の頻度に基づいて単語IDの割り当てを作成
class ID():
    def __init__(self, data): #コンストラクタによる初期化
        self.train_dict = defaultdict(int)  #単語の頻度を保存
        self.id_list = []  #単語と対応するIDを保存
        self.id_dict = dict()  #最終的な単語IDの割り当てを保存
        self.make_id(data)  #make_idメソッドを呼び出して、データを処理

    def make_id(self, data):
        for line in data:
            words = line.translate(table).split() #tableで句読点を削除→単語に分割
            for word in words:
                # if word != "":
                self.train_dict[word] += 1 #単語の頻度を更新
        calc_dict = dict(self.train_dict)
        sort_list = sorted(calc_dict.items(), key=lambda x: x[1], reverse=True)  #頻度の降順にソートされた単語-頻度のペア
        for i, (trg_word, freq) in enumerate(sort_list): #ID割り当て
            if freq >= 2: #頻度が2以上の単語：一意のID（1から開始）を割り当て
                self.id_list.append((trg_word, i+1))  #頻度が2未満の単語：ID 0を割り当て
            else:
                self.id_list.append((trg_word, 0))
        self.id_dict = dict(self.id_list)  #id_list からid_dict(最終的な単語IDの辞書)を作成

    def return_id(self, line): #受け取ったテキスト行を、id_dict に基づいてIDのリストに変換
        one_hot_vec = []
        words = line.strip().split(" ") #行を単語に分割→各単語を id_dict から対応するIDに置換
        for word in words:
            if word in self.id_dict.keys():
                one_hot_vec.append(self.id_dict[word])
            else:
                one_hot_vec.append(0) #単語がid_dict に存在しない場合：0に置換
        return one_hot_vec

#テキストクリーニング：句読点をスペースに置換
table = str.maketrans(string.punctuation, ' '*len(string.punctuation))

#データの読み込み
header_name = ['TITLE', 'CATEGORY']
train_file = 'train.txt'
train_data = pd.read_csv(train_file, header=None,
                         sep='\t', names=header_name)

w2id = ID(train_data['TITLE']) #データセットの'TITLE'列を使用してIDクラスのインスタンス（w2id）を作成
test_vec = w2id.return_id(train_data["TITLE"][10])

#スクリプトがメインプログラムとして実行される場合、id_dictを出力
if __name__ == "__main__":
    # print(test_vec)
    print(w2id.id_dict)
```
