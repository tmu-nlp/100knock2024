```python
#knock71
from torch import nn
import torch

#単一層のパーセプトロンを定義
class SingleLayerPerceptronNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        #線形層 (nn.Linear) を定義し、重みを正規分布で初期化
        self.fc = nn.Linear(input_size, output_size, bias=False)
        nn.init.normal_(self.fc.weight, 0.0, 1.0)

    def forward(self, x):
        x = self.fc(x)  #入力データ x に対して線形変換を適用
        return x

# トレーニングデータの読み込み
X_train = torch.load("X_train.pt")

# モデルのインスタンス化
# input_size を X_train の形状に合わせる
model = SingleLayerPerceptronNetwork(X_train.shape[1], 4)

# 最初のデータポイントの予測
y_hat_1 = torch.softmax(model(X_train[:1]), dim=-1)

# 最初の4つのデータポイントの予測
Y_hat = torch.softmax(model.forward(X_train[:4]), dim=-1)

print("y_hat_1:", y_hat_1)
print("Y_hat:", Y_hat)
```
