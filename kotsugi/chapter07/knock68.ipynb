```python
from importnb import imports

with imports("ipynb"):
  from knock67 import country_vectors, countries
```

```python
from scipy.cluster.hierarchy import linkage

link = linkage(country_vectors, 'ward')
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 15))
dendrogram(link, labels=countries, show_leaf_counts=True)
plt.show()
```
