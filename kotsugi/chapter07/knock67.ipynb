```python
from importnb import imports

with imports("ipynb"):
  from knock60 import model
```

```python
import pandas as pd

df = pd.read_csv("./result_64.csv")
df
```

```python
target_category_A = [
  "capital-common-countries",
  "capital-world",
]

target_category_B = [
  'currency', 
  'gram6-nationality-adjective',
]

countries_A = set(df[df["category"].isin(target_category_A)]["true"].unique())
countries_B = set(df[df["category"].isin(target_category_B)]["vec_1"].unique())

countries = list(countries_A | countries_B)
```

```python
country_vectors = [ model.get_vector(country) for country in countries ]
```

```python
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=5)
idx = kmeans_model.fit(country_vectors)
```

```python
from collections import defaultdict

cluster_labels = kmeans_model.labels_
clusters = defaultdict(list)

for cluster_id, country in zip(cluster_labels, countries):
    clusters[cluster_id].append(country)
```

```python
for cluster in clusters.values():
  print(cluster)
```
