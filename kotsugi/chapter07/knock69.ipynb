```python
import numpy as np
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
countries_tsne = tsne.fit_transform(np.array(country_vectors))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
cmap = plt.get_cmap('Dark2')

plt.scatter(tsne.embedding_[:, 0], tsne.embedding_[:, 1])
for i, ((x, y), name) in enumerate(zip(tsne.embedding_, countries)):
    plt.annotate(name, (x, y), color=cmap(kmeans_model.labels_[i]))
plt.show()
```
