```python
import os
from dotenv import load_dotenv
load_dotenv()

FILE_DIR = os.getenv('FILE_DIR')
```

```python
import pandas as pd

df = pd.read_csv(f"{FILE_DIR}/wordsim353/combined.csv")
df
```

```python
from importnb import imports

with imports("ipynb"):
  from knock60 import model
```

```python
def calc_similarity(df: pd.DataFrame):
  return model.similarity(df["Word 1"], df["Word 2"])
```

```python
df["pred"] = df.apply(calc_similarity, axis=1, result_type="expand")
df
```

```python
df["Human Rank"] = df["Human (mean)"].rank(ascending=False)
df["Pred Rank"] = df["pred"].rank(ascending=False)
df
```

```python
from scipy.stats import spearmanr

correlation, pvalue = spearmanr(df["Human Rank"], df["Pred Rank"])
correlation
```
