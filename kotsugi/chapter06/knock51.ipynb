```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

file_data = []

f = open("./train.txt", "r")
file_data += f.readlines()
f.close()

f = open("./valid.txt", "r")
file_data += f.readlines()
f.close()

f = open("./test.txt", "r")
file_data += f.readlines()
f.close()

file_df = pd.DataFrame(file_data).iloc[:,0].str.split('\t', expand=True).set_axis(["CATEGORY", "TITLE"], axis=1)

vectorizer = TfidfVectorizer(min_df=10)
features = vectorizer.fit_transform(file_df["TITLE"])
feature_names = vectorizer.get_feature_names_out()

feature_df = pd.DataFrame(features.toarray(), columns=feature_names, index=file_df.index).fillna(0)

df = pd.concat([file_df, feature_df], axis=1)
df
```

```python
data_size = df.shape[0]
train_max_idx = int(data_size * 0.80)
valid_max_idx = int((data_size - train_max_idx) * 0.50) + train_max_idx
```

```python
train_df = df.iloc[:train_max_idx]
valid_df = df.iloc[train_max_idx:valid_max_idx]
test_df  = df.iloc[valid_max_idx:-1]
```

```python
train_df.to_csv("./train.feature.txt", index=False)
valid_df.to_csv("./valid.feature.txt", index=False)
test_df.to_csv("./test.feature.txt", index=False)
```
