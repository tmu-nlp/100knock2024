```python
#ファイルの読み込み、中身を配列に格納、集合用の辞書を用意
col1 = open("col1.txt","r")
content = col1.read().splitlines()     
set_col1 = {}

#1行ずつ読み出し、辞書に含まれていなければ出現回数を記録
for name in content:
    if name not in set_col1:
        set_col1[name] = content.count(name)

#作成した辞書を出現回数で降順にソート
sorted_set_col1 = sorted(set_col1.items(),reverse=True,key=lambda x:x[1])

#1タプルずつ表示
for name,count in sorted_set_col1:
    print(count,name)

col1.close()
```
