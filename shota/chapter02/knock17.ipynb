```python
#ファイルの読み込み、1行ずつ配列に格納
col1 = open("col1.txt","r")
content = col1.read().splitlines()     

#配列col1から集合に変換し、表示のために昇順ソート
set_col1 = set(content) 
ls_col1 = sorted(set_col1)

for names in ls_col1:
    print(names)
    
col1.close()
```
