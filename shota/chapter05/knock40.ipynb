```python
class Morph:
  def __init__(self, line):
    surface, other = line.split("\t")
    other = other.split(",")
    self.surface = surface
    self.base = other[6]
    self.pos = other[0]
    self.pos1 = other[1]
 
#文リストの中に、形態素リストが格納されている
sentences = [] #文リスト
morphs = [] #形態素リスト
 
with open("./ai.ja.txt.parsed") as f:
  for line in f:
    if line[0] == "*":                     #係り受けのパラメータを表示している行はスキップ
      continue
    elif line != "EOS\n":                  #文末でなければ　Morphに
      morphs.append(Morph(line))
    else:                                  #EOS（文末）の場合には、morphsをsentenceに追加
      if len(morphs) != 0:
        sentences.append(morphs)
        morphs = []
 
for i in sentences[1]:                     #最初の一文を表示
    print(vars(i))
```
