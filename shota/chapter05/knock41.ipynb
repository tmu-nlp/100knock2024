```python
class Morph:
    def __init__(self, txt):
        col1 = txt.split('\t')
        col2 = col1[1].rstrip().split(',')
        self.surface = col1[0]
        self.base = col2[6]
        self.pos = col2[0]
        self.pos1 = col2[1]

    def __str__(self):
        return 'surface: {}, base: {}, pos: {}, pos1: {}'.format(self.surface, self.base, self.pos, self.pos1)
    #オブジェクトの公式な文字列表現を返すための特殊メソッド、これがないとオブジェクトのクラス名とメモリアドレスを含む文字列が返ってくる
    def __repr__(self):
        return self.surface


class Chunk:
  def __init__(self,morphs,dst):
    self.morphs = morphs
    self.dst = dst
    self.srcs = []
 
def morphs_chunk():
#文リストの中に文節リストが格納されている、文節リストの中に形態素リストが格納されている
    sentences = [] #文リスト
    morphs = [] #形態素リスト
    chunks = [] #文節リスト
    
    with open("./ai.ja.txt.parsed") as f:
        for line in f:
            if line == '人工知能\n':
                continue
            
            if line[0] == "*":                     #係り受けのパラメータを表示している行が現れた時はそこまでを文節として処理
                if len(morphs) != 0:
                    chunks.append(Chunk(morphs,dst))
                morphs = []
                cols = line.split()
                dst = int(cols[2].rstrip('D'))

            elif line != "EOS\n":                  #文末でなければ　Morphに語が追加される
                morphs.append(Morph(line))

            else:                                  #EOS（文末）の場合には、morphsをchunksに追加
                if len(morphs) != 0:
                    chunks.append(Chunk(morphs,dst))
                morphs = []
                if len(chunks) != 0:               #morphsが空でなければ、それをsentenceに追加していく
                    for i,chunk in enumerate(chunks):
                        dst = chunk.dst
                        if dst != -1:
                            chunks[dst].srcs.append(i)
                    sentences.append(chunks)
                chunks = []
    
    return sentences

sentences = morphs_chunk()
for i in sentences[1]:                     #最初の一文を表示
    print(vars(i))
```
