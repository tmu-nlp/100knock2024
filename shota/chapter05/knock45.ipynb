```python
#41で定義した関数を使用するためにimport
from importnb import imports
with imports('ipynb'):
    import knock41 
sentences = knock41.morphs_chunk()

particles = []
for sentence in sentences:                            #文全体から一文ずつ取り出す
    for chunk in sentence:                            #文から文節を取り出す
        for morph in chunk.morphs:                    #文節から単語を取り出す
            if morph.pos == '動詞':                   #単語の品詞が動詞の時
                if len(chunk.srcs) != 0:              #その動詞の係り元の有無
                    modifier = []           #係り元のリスト
                    for src in chunk.srcs:            #係り元について
                        modifier += [x.base for x in sentence[src].morphs if x.pos == '助詞']         #係り元に助詞があればその表層形を追加
                if len(modifier) != 0:                #係り元が存在すれば
                    modifier = sorted(list(set(modifier)))
                    particles.append(morph.base + '\t' + ' '.join(modifier))  #それを出力用の配列に追加
                break

with open ('knock45.txt','w',encoding='utf-8') as f:
    f.write('\n'.join(particles))

print('\n'.join(particles))
```
